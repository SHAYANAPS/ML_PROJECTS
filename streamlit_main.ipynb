Outcome = 1:  presence of diabetes.

Outcome = 0:  absence of diabetes.

```python
df.groupby("Outcome").mean()
```

```python
X=df.drop(columns="Outcome",axis=1)
Y=df["Outcome"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=2) 
# to make sure that the selection of data for trainign contains equal proportion of both classes(0 and 1)
```

```python
from sklearn.svm import SVC 
```

```python
classifier=SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
training_data=accuracy_score(Y_pred,Y_test)
```

```python
print("Accuracy of SVC:",training_data)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(multi_class="ovr",solver="saga",max_iter=10000)
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=50)
```

```python
rf.fit(X_train,Y_train)
```

```python
rf.predict(X_test)
```

```python
rf.score(X_test,Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=5)
```

```python
dt.fit(X_train,Y_train)
```
